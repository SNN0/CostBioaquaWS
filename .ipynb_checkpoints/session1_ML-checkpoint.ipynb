```python
# ==== CODE CELL 1 ====
# Libraries & global settings
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import joblib

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, LabelBinarizer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
from boruta import BorutaPy

from sklearn.metrics import (classification_report, confusion_matrix, ConfusionMatrixDisplay,
                             roc_curve, auc, RocCurveDisplay, precision_recall_curve, 
                             average_precision_score, roc_auc_score)

import shap
from itertools import cycle

# Set global styles and configurations
sns.set_style('whitegrid')
warnings.filterwarnings('ignore')
print("Libraries loaded. We can start the analysis.")
```

```python
# =============================================================================
# PART 1: DATA LOADING, CLEANING, AND SPLITTING
# =============================================================================

# --- CELL 2: Load Data and Clean Column Names ---
# Description:
# We are loading the data and fixing issues in the column names.
# We will STRICTLY check the state of the data before and after each step.

df = pd.read_csv('data/WQD.csv')
print("--- DATA BEFORE CLEANING ---")
print("Original Column Names:", df.columns.tolist())
display(df.head())

def clean_col_names(df_to_clean):
    """Cleans column names by removing special characters and standardizing format."""
    new_cols = []
    for col in df_to_clean.columns:
        clean_col = col.replace('`', '').strip()
        clean_col = re.sub(r'\s*\([^)]*\)', '', clean_col)
        clean_col = re.sub(r'[\s/]+', '_', clean_col)
        new_cols.append(clean_col)
    df_to_clean.columns = new_cols
    return df_to_clean

df = clean_col_names(df)
print("\n--- DATA AFTER CLEANING ---")
print("New Column Names:", df.columns.tolist())
display(df.head())
```

```python
# --- CELL 3: Split Data into Main and Inference Sets ---
# Description:
# We are setting aside 10% of our data for a final "live test" (inference).
# All standard procedures will be performed on the remaining 90%.

quality_map = {0: 'Excellent', 1: 'Good', 2: 'Poor'}
df['Water_Quality_Label'] = df['Water_Quality'].map(quality_map)

main_df, inference_df = train_test_split(df, test_size=0.10, random_state=42, stratify=df['Water_Quality'])

print("\n--- Data Set Split in Two ---")
print(f"Size of the main data for training and testing (main_df): {main_df.shape}")
print(f"Size of the reserved data for live testing (inference_df): {inference_df.shape}")
```

```python
# =============================================================================
# PART 2: EXPLORATORY DATA ANALYSIS (EDA)
# =============================================================================

# --- CELL 4: General Data Overview ---
# Description:
# We examine the structure of our cleaned dataset using .info() and .describe().

print("General Information About the Data Set:")
main_df.info()

print("\nStatistical Summary of the Data Set:")
display(main_df.describe().T)
```

```python
# --- CELL 5: Target Variable and Correlation Analysis ---
# Description: We analyze the distribution of our target variable and examine the
# relationship between features using a heatmap.

print("'Water_Quality_Label' column was added to the dataset:")
display(main_df[['Water_Quality', 'Water_Quality_Label']].head())

plt.figure(figsize=(10, 6))
sns.countplot(x='Water_Quality_Label', data=main_df, order=['Excellent', 'Good', 'Poor'], palette='viridis')
plt.title('Distribution of Water Quality Classes', fontsize=16)
plt.xlabel('Water Quality')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(14, 10))
sns.heatmap(main_df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Features', fontsize=16)
plt.show()
```

```python
# --- CELL 6: Feature Distributions and Class-Based Comparison ---
# Description: We gain a deeper understanding of the data by examining feature
# distributions with histograms and box plots.

main_df.drop(['Water_Quality', 'Water_Quality_Label'], axis=1).hist(figsize=(16, 12), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Distribution of All Features (Histograms)', fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

features_to_plot = ['Ammonia', 'DO', 'pH', 'Turbidity']
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Variation of Key Features by Water Quality Class', fontsize=20)
for i, feature in enumerate(features_to_plot):
    row, col = i // 2, i % 2
    sns.boxplot(x='Water_Quality_Label', y=feature, data=main_df, ax=axes[row, col], order=['Excellent', 'Good', 'Poor'])
    axes[row, col].set_title(f'{feature} Distribution', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# =============================================================================
# PART 3: DATA PREPROCESSING
# =============================================================================

# --- CELL 7: Data Scaling Methods and Formulas ---
# Description:
# Most machine learning models are negatively affected by features being on different scales.
# For example, if one feature ranges from 0-1 and another from 1000-5000, the model
# might tend to give more "weight" to the feature with larger values. We perform scaling
# to prevent this bias. The two most popular methods are:

# 1. Standardization (StandardScaler):
#    Transforms the data to have a mean of 0 and a standard deviation of 1.
#    It is more robust to outliers and is generally the default choice.
#    Formula: $$z = \frac{x - \mu}{\sigma}$$
#    ($\mu$: mean, $\sigma$: standard deviation)

# 2. Normalization (MinMaxScaler):
#    Rescales the data to a fixed range, usually 0 to 1.
#    The minimum value of the feature becomes 0, and the maximum value becomes 1.
#    Formula: $$X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

# --- CELL 8: Understanding Methods with a Mini Example ---
sample_data = np.array([[2, 100], [5, 500], [8, 900]], dtype=float)
sample_df = pd.DataFrame(sample_data, columns=['Feature_A', 'Feature_B'])
print("--- MINI SAMPLE DATA BEFORE SCALING ---"); display(sample_df)

# Standardization Application and Proof
print("\n" + " 1. Standardization (StandardScaler) Application ".center(70, "-"))
ss_scaler = StandardScaler()
ss_scaled = ss_scaler.fit_transform(sample_data)
ss_scaled_df = pd.DataFrame(ss_scaled, columns=['Feature_A', 'Feature_B'])
print("AFTER Standardization:"); display(ss_scaled_df)
print("PROOF (Mean and Standard Deviation):")
print(f"  Means (should be close to 0): {ss_scaled.mean(axis=0)}")
print(f"  Standard Deviations (should be 1): {ss_scaled.std(axis=0)}")

# Normalization Application and Proof
print("\n" + " 2. Normalization (MinMaxScaler) Application ".center(70, "-"))
mms_scaler = MinMaxScaler()
mms_scaled = mms_scaler.fit_transform(sample_data)
mms_scaled_df = pd.DataFrame(mms_scaled, columns=['Feature_A', 'Feature_B'])
print("AFTER Normalization:"); display(mms_scaled_df)
print("PROOF (Minimum and Maximum Values):")
print(f"  Minimum Values (should be 0): {mms_scaled.min(axis=0)}")
print(f"  Maximum Values (should be 1): {mms_scaled.max(axis=0)}")
```

```python
# --- CELL 9: Data Splitting and Scaling ---
# Description: We show the output of each data splitting and scaling step.
X = main_df.drop(['Water_Quality', 'Water_Quality_Label'], axis=1)
y = main_df['Water_Quality']

# Step 1: Split the Data Set
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

print("--- Step 1: Data Set Split into Training and Test Sets ---")
print(f"Training set (X_train_raw) shape: {X_train_raw.shape}")
print(f"Test set (X_test_raw) shape:   {X_test_raw.shape}")
print(f"Training target (y_train) shape: {y_train.shape}")
print(f"Test target (y_test) shape:   {y_test.shape}")

# Step 2: Scale the Features
print("\n--- Step 2: Scaling Features ---")
print("First 5 rows of X_train_raw BEFORE scaling:")
display(X_train_raw.head())

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)

X_train_scaled_df = pd.DataFrame(X_train, columns=X.columns)
print("\nFirst 5 rows of X_train AFTER scaling:")
display(X_train_scaled_df.head())
```

```python
# =============================================================================
# PART 4: THE CRITICAL IMPORTANCE OF SCALING & BASELINE MODELS
# =============================================================================

# --- CELL 10: SVM Performance WITHOUT Scaling ---
# Description: This cell is here to prove the importance of data preprocessing.
print("--- SVM Results with Unscaled (Raw) Data ---")
svm_unscaled = SVC(random_state=42).fit(X_train_raw, y_train)
y_pred_unscaled = svm_unscaled.predict(X_test_raw)
print("â¼ï¸ NOTE: These low scores obtained without scaling prove that preprocessing is vital.")
print(classification_report(y_test, y_pred_unscaled, target_names=quality_map.values()))
```

```python
# --- CELL 11: Training Models with Default Parameters ---
# Description: We train our models "out of the box" without any hyperparameter tuning.
baseline_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Support Vector Machine (SVM)": SVC(random_state=42, probability=True),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')
}
print(" PERFORMANCE OF BASELINE MODELS ".center(70, "="))
for name, model in baseline_models.items():
    print(f"\n--- {name} Model (Default Settings) ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred, target_names=quality_map.values()))
```

```python
# =============================================================================
# PART 5: HYPERPARAMETER OPTIMIZATION (GRIDSEARCHCV)
# =============================================================================

# --- CELL 12: Searching for the Best Parameters with GridSearchCV ---
# Description: We search for the best hyperparameters to improve baseline performance.
print(" HYPERPARAMETER OPTIMIZATION STARTED ".center(70, "="))

# 1. Search for best parameters for Random Forest
print("\nSearching for best parameters for RandomForest...")
param_grid_rf = {'n_estimators': [100, 200], 'max_depth': [10, 20], 'min_samples_leaf': [1, 2]}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, n_jobs=-1, verbose=1)
grid_rf.fit(X_train, y_train)
best_rf = grid_rf.best_estimator_
print(f"Best RF Parameters: {grid_rf.best_params_}")

# 2. Search for best parameters for SVM
print("\nSearching for best parameters for SVM...")
param_grid_svm = {'C': [10, 100], 'gamma': ['scale', 0.01]}
grid_svm = GridSearchCV(SVC(random_state=42, probability=True), param_grid_svm, cv=5, n_jobs=-1, verbose=1)
grid_svm.fit(X_train, y_train)
best_svm = grid_svm.best_estimator_
print(f"Best SVM Parameters: {grid_svm.best_params_}")

# 3. Search for best parameters for XGBoost
print("\nSearching for best parameters for XGBoost...")
param_grid_xgb = {'n_estimators': [100, 200], 'max_depth': [3, 5], 'learning_rate': [0.1, 0.2]}
grid_xgb = GridSearchCV(XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'), param_grid_xgb, cv=5, n_jobs=-1, verbose=1)
grid_xgb.fit(X_train, y_train)
best_xgb = grid_xgb.best_estimator_
print(f"Best XGBoost Parameters: {grid_xgb.best_params_}")
```

```python
# =============================================================================
# PART 6: FINAL MODEL COMPARISON
# =============================================================================

# --- CELL 13: Comparing Baseline and Optimized Models ---
# Description: We compare baseline and tuned models to see the impact of optimization.
final_models = {
    "1. Random Forest (Baseline)": baseline_models["Random Forest"],
    "2. Random Forest (Optimized)": best_rf,
    "3. XGBoost (Baseline)": baseline_models["XGBoost"],
    "4. XGBoost (Optimized)": best_xgb,
    "5. SVM (Baseline)": baseline_models["Support Vector Machine (SVM)"],
    "6. SVM (Optimized)": best_svm
}
print("\n" + " COMPARING FINAL MODELS ".center(70, "="))
for name, model in final_models.items():
    print(f"\n--- {name} ---")
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred, target_names=quality_map.values()))

final_best_model = best_xgb # We select the best model
```

```python
# --- CELL 14: Confusion Matrices for Final Models ---
models_to_plot = {
    "Random Forest (Baseline)": baseline_models["Random Forest"],
    "Random Forest (Optimized)": best_rf,
    "XGBoost (Baseline)": baseline_models["XGBoost"],
    "XGBoost (Optimized)": best_xgb
}
fig, axes = plt.subplots(2, 2, figsize=(14, 12)); fig.suptitle('Comparison of Baseline and Optimized Models', fontsize=20); axes = axes.flatten()
for i, (name, model) in enumerate(models_to_plot.items()):
    ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, display_labels=quality_map.values(), cmap=plt.cm.Blues, ax=axes[i]); axes[i].set_title(name)
plt.tight_layout(rect=[0, 0, 1, 0.96]); plt.show()
```

```python
# --- CELL 15: ROC Curve and AUC Score ---
# Description: The ROC Curve shows how well a model can distinguish between classes.
# The closer the Area Under the Curve (AUC) is to 1, the more successful the model.

# Binarize the output for multiclass ROC analysis
lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)
n_classes = y_test_bin.shape[1]

plt.figure(figsize=(12, 10))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple'])

for (name, model), color in zip(final_models.items(), colors):
    y_score = model.predict_proba(X_test)
    
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute macro-average ROC curve and ROC area
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= n_classes
    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

    plt.plot(fpr["macro"], tpr["macro"], color=color, lw=2,
             label=f'{name} (macro-avg AUC = {roc_auc["macro"]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison for Models (Macro-Average)', fontsize=16)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# =============================================================================
# PART 7: MODEL INTERPRETABILITY (SHAP)
# =============================================================================

# --- CELL 16: Understanding Model Decisions with SHAP ---
# Description: SHAP is a modern method that shows how much and in what direction
# each feature affects a model's prediction.
print("--- Interpreting the Best Model (XGBoost) with SHAP ---")
explainer = shap.TreeExplainer(final_best_model)
# Create a modern SHAP Explanation object, which is easier and safer to work with
shap_explanation = explainer(pd.DataFrame(X_test, columns=X.columns))

print("\nSHAP Values Calculated. Summary Plot:")

# Global Feature Importance
# The summary_plot works directly with the new explanation object
shap.summary_plot(shap_explanation, plot_type="bar", class_names=list(quality_map.values()))


# --- Explanation of a single prediction (the first test sample) ---
print("\n--- Prediction Explanation for the First Test Sample ---")
true_label_index = y_test.iloc[0]
predicted_label_index = final_best_model.predict(X_test[0:1])[0]

print(f"Model prediction: {quality_map[predicted_label_index]}, Actual value: {quality_map[true_label_index]}")
shap.initjs()

# FIX: Use the Explanation object for plotting to avoid dimension errors.
# We will create a plot explaining the model's prediction for the 'Excellent' class (index 0) for the first sample.
shap.force_plot(shap_explanation[0, :, 0])
```

```python
# =============================================================================
# PART 8: ADVANCED FEATURE SELECTION WITH BORUTA
# =============================================================================

# --- CELL 17: Running the Boruta Algorithm ---
# Description: Boruta finds statistically significant features.
print("--- Feature Selection with Boruta Initiated (This may take a while) ---")
rf_boruta = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5)
boruta_selector = BorutaPy(rf_boruta, n_estimators='auto', verbose=2, random_state=42)
# Boruta expects numpy arrays
boruta_selector.fit(np.array(X_train), np.array(y_train))

print("\n--- Boruta Analysis Completed ---")
selected_boruta_features = X.columns[boruta_selector.support_].tolist()
print(f"Out of {len(X.columns)} total features, {len(selected_boruta_features)} were confirmed as 'Important' by Boruta:")
print("Confirmed Features:", selected_boruta_features)

# Show the potential reduction
print(f"\nData can be reduced from {X_train.shape[1]} features to {len(selected_boruta_features)} features after Boruta selection.")
```

```python
# =============================================================================
# PART 9: FINAL INTERPRETATION - FEATURE IMPORTANCE
# =============================================================================

# --- CELL 18: Feature Importance Ranking According to the Best Model ---
# Description: Let's visualize the most important parameters affecting water quality,
# according to our best-performing optimized Random Forest model.

best_model_for_importance = best_rf # Using the optimized RF model
importances = best_model_for_importance.feature_importances_
feature_importance_df = pd.DataFrame({'feature': X.columns, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='importance', y='feature', data=feature_importance_df, palette='rocket')
plt.title('Feature Importance Ranking for Water Quality Prediction (Optimized Random Forest)', fontsize=16)
plt.show()
```

```python
# =============================================================================
# PART 10: SAVING AND LOADING THE MODEL
# =============================================================================

# --- CELL 19: Saving the Best Model and Scaler ---
# Description: In a real-world scenario, we save our trained model and scaler
# to disk for later use.
model_file = 'models/water_quality_xgb_model.joblib'
scaler_file = 'models/water_quality_scaler.joblib'

joblib.dump(final_best_model, model_file)
joblib.dump(scaler, scaler_file)
print(f"Best model saved as '{model_file}'.")
print(f"Scaler object saved as '{scaler_file}'.")
```

```python
# --- CELL 20: Loading the Saved Model ---
loaded_model = joblib.load(model_file)
loaded_scaler = joblib.load(scaler_file)
print("\nModel and Scaler successfully loaded from disk.")
print("Loaded model:", loaded_model)
```

```python
# =============================================================================
# PART 11: LIVE PREDICTION DEMO (INFERENCE)
# =============================================================================

# --- CELL 21: Selecting Samples from the Inference Set and Predicting ---
# Description: We will select 3 samples from the inference set we initially created
# and have our best model predict their quality.
sample_to_predict = inference_df.sample(3, random_state=73)
X_predict = sample_to_predict.drop(['Water_Quality', 'Water_Quality_Label'], axis=1)
y_true = sample_to_predict['Water_Quality']

print("--- 3 Samples Selected for Live Prediction ---"); display(sample_to_predict)

# Use the loaded scaler and model
X_predict_scaled = loaded_scaler.transform(X_predict)
print("\n--- Selected Samples Scaled for the Model ---"); display(pd.DataFrame(X_predict_scaled, columns=X.columns))

predictions = loaded_model.predict(X_predict_scaled)
prediction_labels = [quality_map[p] for p in predictions]
results = pd.DataFrame({'Actual Value': y_true.map(quality_map).values, 'Model Prediction': prediction_labels})
results['Result'] = np.where(results['Actual Value'] == results['Model Prediction'], 'â CORRECT', 'â INCORRECT')

print("\n" + " LIVE PREDICTION RESULTS ".center(70, "=")); display(results)
```

```python
# =============================================================================
# PART 9: RETRAINING WITH THE MOST IMPORTANT FEATURES
# =============================================================================

# --- CELL 22: Selecting the Top 6 Features and Re-evaluating Models ---
# Description:
# In this section, we simplify the dataset by using the top 6 features identified
# as most important by the Random Forest model. Our goal is to see how successful
# the models can be with only these critical features and to understand if there is
# a significant drop in performance.

print(" PART 9: RETRAINING WITH THE TOP 6 FEATURES ".center(70, "="))

# Step 1: Get the model and ranking that determines feature importance
# (We used similar code previously in the feature importance section)
model_for_importance = best_rf
importances = model_for_importance.feature_importances_
feature_importance_df = pd.DataFrame({'feature': X.columns, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

# Step 2: Get the names of the top 6 most important features into a list
top_6_features = feature_importance_df['feature'].head(6).tolist()
print("\nTop 6 most important features selected according to Random Forest:")
print(top_6_features)

# Step 3: Filter our training and test sets according to these 6 features
# IMPORTANT: We are using the raw data (e.g., X_train_raw) before it was scaled.
X_train_top6_raw = X_train_raw[top_6_features]
X_test_top6_raw = X_test_raw[top_6_features]

print(f"\nDatasets were reduced from {X_train_raw.shape[1]} features to {X_train_top6_raw.shape[1]} features.")
print("First 5 rows of the filtered training data:")
display(X_train_top6_raw.head())

# Step 4: Scale our new, simplified dataset
scaler_top6 = StandardScaler()
X_train_top6_scaled = scaler_top6.fit_transform(X_train_top6_raw)
X_test_top6_scaled = scaler_top6.transform(X_test_top6_raw)
print("\nFiltered data has been rescaled.")

# Step 5: Retrain the baseline models with the top 6 features and compare the results
print("\n" + " RE-TRAINING MODELS WITH TOP 6 FEATURES ".center(70, "-"))
for name, model in baseline_models.items():
    print(f"\n--- {name} Model (Top 6 Features) ---")
    model.fit(X_train_top6_scaled, y_train)
    y_pred_top6 = model.predict(X_test_top6_scaled)
    print(classification_report(y_test, y_pred_top6, target_names=quality_map.values()))

print("\nAnalysis complete. The results can now be compared with the baseline model results that used all features.")
```
